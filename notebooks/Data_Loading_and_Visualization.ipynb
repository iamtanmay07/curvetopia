```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def load_polyline(csv_path):
    """Load polyline data from a CSV file and return as a DataFrame."""
    return pd.read_csv(csv_path)

def plot_polyline(polyline):
    """Plot a polyline from a DataFrame."""
    plt.figure(figsize=(10, 10))
    plt.plot(polyline['x'], polyline['y'], marker='o', linestyle='-')
    plt.title("Visualization of the Polyline")
    plt.xlabel("X coordinate")
    plt.ylabel("Y coordinate")
    plt.grid(True)
    plt.axis('equal')
    plt.show()
```

```python
# Update the path according to your project structure
csv_path = '../data/raw/sample_polyline.csv'
polyline = load_polyline(csv_path)
plot_polyline(polyline)
```
